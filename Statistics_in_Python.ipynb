```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import bernoulli

def bernouill_dist(p,n):

    '''Функция создает экземпляр класса Бернули, считает его статистики
      и визуализирует sample.

      p - вероятность успеха, n - размер выборки

    '''
    dist = bernoulli(p)

    # 'mv' означает mean (среднее) и variance (дисперсия)
    mean, var = dist.stats(moments='mv')

    print(f"Теоретическое среднее: {mean}")
    print(f"Теоретическая дисперсия: {var}")

    data = dist.rvs(size=1000)

    plt.figure(figsize=(6, 4))
    # Считаем уникальные значения (сколько 0 и сколько 1)
    unique, counts = np.unique(data, return_counts=True)
    plt.bar(unique, counts, color=['salmon', 'skyblue'], tick_label=['0 (Неудача)', '1 (Успех)'])
    plt.title(f"Симуляция Бернулли (p={p})")
    plt.ylabel("Количество выпадений")
    plt.show()

bernouill_dist(0.4,100)
```

```python
from scipy.stats import binom

def binomial_dist_vis(n,p):

    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    plt.bar(x, pmf)
    plt.title(f"Биномиальное распределение (n={n}, p={p})")
    plt.show()
```

```python
binomial_dist_vis(1000,0.5)
```

```python
binomial_dist_vis(1000,0.8)
```

```python
from scipy.stats import geom

def geom_dist_vis(p,k):

    rv = geom(p)
    x = np.arange(1, k+1)

    pmf_values = rv.pmf(x)

    plt.figure(figsize=(10, 5))
    plt.bar(x, pmf_values, color='cornflowerblue', alpha=0.7)

    for i, v in enumerate(pmf_values):
        plt.text(x[i], v + 0.005, f"{v:.2f}", ha='center')

    plt.title(f"Геометрическое распределение (p={p}")
    plt.xlabel("Номер попытки, на которой случился успех")
    plt.ylabel("Вероятность")
    plt.xticks(x)
    plt.show()

geom_dist_vis(0.3, 11)
```

```python
from scipy.stats import hypergeom

def  hyper_dist(N,D,n):

    # В scipy параметры называются (M, n, N) -> (Всего, Успешных, Выборка)
    rv = hypergeom(M=N, n=D, N=n)

    x = np.arange(0, n+1)
    pmf_dogs = rv.pmf(x)

    plt.figure(figsize=(8, 5))
    plt.bar(x, pmf_dogs, color='purple', alpha=0.7)
    plt.title(f"Гипергеометрическое распределение с параметрами {N, D, n}")
    plt.xlabel("Количество тузов")
    plt.ylabel("Вероятность")
    plt.show()

hyper_dist(36,4,9)
```

```python
from scipy.stats import poisson

def poisson_dist(l):

    x = np.arange(0, 15)
    pmf = poisson.pmf(x, l)

    plt.bar(x, pmf, color='teal', alpha=0.7)
    plt.title(f"Распределение Пуассона ($\\lambda={l}$)")
    plt.xlabel("Количество событий")
    plt.show()

    data = poisson.rvs(l, size=10000)
    print(f"Mean: {data.mean():.2f}, Var: {data.var():.2f}")

poisson_dist(10)
```
