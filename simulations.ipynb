```python
n = 52
r = 1000
def fixed_point_permutations(n):

    flip_num = np.arange(1, n+1)

    deck_values = np.random.permutation(n)

    return  np.sum(flip_num == deck_values)

results = [fixed_point_permutations(n) for _ in range(r)]

print(f"Вероятность хотя бы одного совпадения: {np.mean(np.array(results) > 0)}")
```

```python
k = 23
r = 1000

def birthday_match(k):

    days = np.random.randint(1,366,size=k)

    return len(np.unique(days)) < k

sim = [birthday_match(k) for _ in range(r)]

print(f'Вероятность совпадения: {np.mean(sim)}')
```
